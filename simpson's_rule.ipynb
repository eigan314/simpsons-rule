```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#def function
def f(x):
   return 1/(20 + x**2)

x = np.linspace(-10, 10, 100)

plt.plot(x, f(x), color='blue')

plt.show()
```

```python
n = int(input("Enter the number of intervals = "))    #number of intervals n
x0 = a = float(input("Enter the lower limit = "))     #lower limit
xn = b = float(input("Enter the upper limit = "))     #upper limit
```

```python
#find h

h = (b - a)/n
print("h = ", h)

#find sum of f(x1) + f(x2) +....f(xn-1))
sum_f_odd = 0
sum_f_even = 0

for i in range(1, n):
  if i%2 == 0 and i < n-1:
    sum_f_even += f(x0 + i*h)
  else:
    sum_f_odd += f(x0 + i*h)
```

```python
#simpson's formula

S = (h/3)*(f(a) + f(b) + + 2*(sum_f_even) + 4*(sum_f_odd))
print("Area under the curve = ", S)
```
